### Data Loading

```python
from pathlib import Path

import polars as pl

pl.Config.set_tbl_cols(-1)
pl.Config.set_tbl_rows(50)
pl.Config.set_fmt_str_lengths(120)

DATA_DIR = Path("../../data")
JUDGMENTS_METADATA_PATH = DATA_DIR / "art_6_judgments_metadata_processed.json"
```

```python
judgments_metadata = pl.read_ndjson(JUDGMENTS_METADATA_PATH, infer_schema_length=None)

print(f"Judgments metadata shape: {judgments_metadata.shape}")
```

```python
metadata_columns = pl.DataFrame(
            {
                "dataset": ["judgments"] * len(judgments_metadata.columns),
                "column": judgments_metadata.columns,
                "dtype": [str(dtype) for dtype in judgments_metadata.dtypes],
            }
)

metadata_columns
```
